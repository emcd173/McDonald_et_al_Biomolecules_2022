```python
# begin by impoprting some useful libraries 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import math 
from sklearn.metrics import r2_score
from scipy import stats
```

```python
WT_cl_path13 = '../refinement_optimization/closed/frag_leng_opt/13/'
WT_cl_path7 = '../refinement_optimization/closed/frag_leng_opt/7/'

WT_op_path13 = '../refinement_optimization/open/frag_leng_opt/13/'
WT_op_path7 = '../refinement_optimization/open/frag_leng_opt/7/'
```

## Closed State

#### import data

```python
WT_rmsd_cl7 = []
for i in range(100): 
    fnm = WT_cl_path7 +'struct_RMSD/' + '5uak_WT_struct_rmsd_{}.pdb'.format(i)
    x = np.genfromtxt(fnm, dtype = ['U15', 'float64'],)
    x.shape = [1,]
    WT_rmsd_cl7.append(float(x[0][1]))

WT_score_cl7 = []
for i in range(100):
    fnm = WT_cl_path7 +'scores/' + 'score_{}.sc'.format(i)
    x = np.genfromtxt(fnm, dtype = ['U15', 'float64'], skip_header = 1 )
    WT_score_cl7.append(float(x[1][1]))
```

```python
WT_rmsd_cl13 = []
for i in range(100): 
    fnm = WT_cl_path13 +'struct_RMSD/' + '5uak_WT_struct_rmsd_{}.pdb'.format(i)
    x = np.genfromtxt(fnm, dtype = ['U15', 'float64'],)
    x.shape = [1,]
    WT_rmsd_cl13.append(float(x[0][1]))

WT_score_cl13 = []
for i in range(100):
#for i in np.concatenate([range(0,400), range(402,735), range(736,752), range(753, 1000)]): 
    fnm = WT_cl_path13 +'scores/' + 'score_{}.sc'.format(i)
    x = np.genfromtxt(fnm, dtype = ['U15', 'float64'], skip_header = 1 )
    WT_score_cl13.append(float(x[1][1]))
```

#### plot 

```python
x = WT_rmsd_cl7
y = WT_score_cl7
x1 = WT_rmsd_cl13
y1= WT_score_cl13

font = 30
#print(min(x))
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

fig, ax = plt.subplots()
plt.tick_params(axis='both', which='major', labelsize=font-10)


fig.set_figheight(12)
fig.set_figwidth(24)

ax.scatter(x,y, s=250, c = 'deepskyblue', marker = 'o', edgecolors='k')
ax.scatter(x1,y1, s=250, c = 'red', marker = 'o', edgecolors='k')
#ax.scatter(0,min(y1), s=100, c='lime', marker= 'o', edgecolors='k')
#ax.scatter(x1, y1, s=100, c = 'lightskyblue', marker = 'o', )
#ax.plot(x, [i*slope +intercept for i in x], '--', c= 'r' )
#ax.plot([min(x), max(x)], [min(y), max(y)],
         #'--', color = 'k') 

ax.set_xlabel('RMSD', fontsize =font)
ax.set_ylabel('Score',fontsize =font)
ax.ticklabel_format(labelssize =font )
ax.set_xlim(0,2.1)
ax.xaxis.labelpad = 30
ax.yaxis.labelpad = 30
#ax.text(max(x)-1 , min(y)+50 , 'r = {:.3f}'.format(r_value), color='k', 
 
    #bbox=dict(facecolor='none', edgecolor='k', boxstyle='square,pad=1'))
    
fig.savefig('../figures/5uak_fragment_insertion_score_vs_rmsd.svg', format='svg')
```
